```python
#Bài 5:
def doc_so(n):
    hang_don_vi = ["", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]
    hang_chuc = ["", "mười", "hai mươi", "ba mươi", "bốn mươi", "năm mươi",
                 "sáu mươi", "bảy mươi", "tám mươi", "chín mươi"]

    if n < 10:   #số 1 chữ số
        return hang_don_vi[n]

    chuc = n // 10
    donvi = n % 10

    if donvi == 0:
        return hang_chuc[chuc]
    elif donvi == 1 and chuc > 1:
        return hang_chuc[chuc] + " mốt"
    elif donvi == 5 and chuc > 0:
        return hang_chuc[chuc] + " lăm"
    else:
        return hang_chuc[chuc] + " " + hang_don_vi[donvi]


#Nhập số và in ra cách đọc
n = int(input("Nhập số có tối đa 2 chữ số: "))
if 0 <= n < 100:
    print("Cách đọc:", doc_so(n))
else:
    print("Vui lòng nhập số từ 0 đến 99")
```

```python
#Bài 6:
import math

#Nhập hệ số
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

#Giải phương trình
if a == 0:  #Trường hợp trở thành bậc nhất
    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        print("Phương trình có một nghiệm x =", -c/b)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
```

```python
#Bài 16:
def nam_nhuan(nam):
    #Kiểm tra năm nhuận
    return (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0)

def ngay_hople(ngay, th, nam):
    #Kiểm tra ngày tháng năm hợp lệ
    #Tháng có số ngày tương ứng
    ngay_trongthang = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    if nam <= 0 or th <= 0 or th > 12 or ngay <= 0:
        return False

    #Nếu tháng 2 và năm nhuận
    if th == 2 and nam_nhuan(nam):
        max_day = 29
    else:
        max_day = ngay_trongthang[th]

    return ngay <= max_day

NGAY = int(input("Nhập ngày: "))
THANG = int(input("Nhập tháng: "))
NAM = int(input("Nhập năm: "))

if ngay_hople(NGAY, THANG, NAM):
    print("Ngày tháng năm hợp lệ.")
else:
    print("Ngày tháng năm KHÔNG hợp lệ.")
```

```python
#Bài 18 - câu a:
import math
#Hàm kiểm tra số nguyên tố
def sont(x):
    if x < 2:
        return False
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            return False
    return True

#Nhập số n
n = int(input("Nhập n: "))
uocnt = []
for i in range(2, n + 1):
    if n % i == 0 and sont(i):
        uocnt.append(i)
print("Các ước số nguyên tố khác nhau của n:", uocnt)
print("Số lượng =", len(uocnt))
```

```python
#Bài 18 - câu c:
#Hàm nhập n từ bàn phím
def nhap_n():
    while True:
        try:
            n = int(input("Nhập n (số tự nhiên ≥ 1): "))
            if n >= 1:
                return n
            print("n phải ≥ 1, nhập lại!")
        except ValueError:
            print("Vui lòng nhập số nguyên hợp lệ!")

#Hàm tính tổng 12 + 22 + 32 + ... + n2
def tong(n: int) -> int:
    s = 0
    for i in range(1, n + 1):
        s += i * 10 + 2   #số có dạng i2 (ví dụ: 12, 22, 32, ...)
    return s

#Hàm xuất kết quả
def xuat_kq(n, s):
    print(f"Tổng 12 + 22 + 32 + ... + {n}2 = {s}")

#Hàm main: điều phối các bước
def main():
    n = nhap_n()      #nhập n
    s = tong(n)       #tính tổng
    xuat_kq(n, s)     #in kết quả

#Chạy chương trình
if __name__ == "__main__":
    main()
```

```python
#Bài 19 - câu c:
import math
#Đổi giây ra giờ, phút, giây
def doi_giay(s):
    gio = s // 3600
    s = s % 3600
    phut = s // 60
    giay = s % 60
    return gio, phut, giay
#Nhập số giây
s = int(input("Nhập số giây: "))
gio, phut, giay = doi_giay(s)
print(f"{s} giây = {gio} giờ {phut} phút {giay} giây")
```

```python
import math
#Kiểm tra số nguyên tố
def snt(x: int) -> bool:
    if x < 2:
        return False
    if x == 2:
        return True
    if x % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(x)) + 1, 2):
        if x % i == 0:
            return False
    return True

#Kiểm tra số nguyến tố lớn nhất trong dàn ước chung
def f(m: int, n: int) -> int:
    gcd = math.gcd(m, n)
    snt_lonI = 0
    for i in range(2, gcd + 1):
        if gcd % i == 0 and snt(i):
            snt_lonI = i
    return snt_lonI

#Nhập
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))
print("Số nguyên tố lớn nhất và là ước chung của",m,"và",n,"là",f(m,n))
```
